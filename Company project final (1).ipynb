```python
from scipy.stats import ttest_ind

# Hypothesis Testing: Compare Base MSRP between Electric Vehicle Types
ev_types = df['Electric Vehicle Type'].unique()
for ev_type in ev_types:
    subset = df[df['Electric Vehicle Type'] == ev_type]
    print(f'T-Test for {ev_type} vs. Others:')
    stat, p_value = ttest_ind(subset['Base MSRP'], df[df['Electric Vehicle Type'] != ev_type]['Base MSRP'])
    print(f'T-Statistic: {stat}, p-Value: {p_value}\n')
```

The extremely high positive t-statistic indicates that the mean Base MSRP for Battery Electric Vehicles (BEV) is significantly higher than the mean Base MSRP for the combined group of other electric vehicle types.
The p-value of 0.0 suggests strong statistical evidence against the null hypothesis. It essentially means that the observed difference in Base MSRP between BEV and others is not due to random chance but is a real and significant difference.

The extremely high negative t-statistic indicates that the mean Base MSRP for Plug-in Hybrid Electric Vehicles (PHEV) is significantly lower than the mean Base MSRP for the combined group of other electric vehicle types.
Similar to BEV, the p-value of 0.0 provides strong statistical evidence against the null hypothesis. It indicates that the observed difference in Base MSRP between PHEV and others is not random but is a real and significant difference.

### Analyzing Feature Relevance and Importance:

```python
# Correlation matrix
correlation_matrix = df.corr()

# Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.show()
```

### Feature Selection and Engineering

```python
# Drop irrelevant columns
columns_to_drop = ['Identifier', 'Vehicle ID']
df = df.drop(columns_to_drop, axis=1)
```

```python
df = pd.get_dummies(df, columns=['City', 'Make', 'Model', 'Electric Vehicle Type',
                                 'Clean Alternative Fuel Vehicle (CAFV) Eligibility', 
                                 'Vehicle Location', 'Electric Utility'])
```

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Performing feature selection using random forest
X = df.drop('Base MSRP', axis=1)
y = df['Base MSRP']
selector = SelectFromModel(RandomForestRegressor())
selector.fit(X, y)
selected_features = X.columns[selector.get_support()]
```

```python
# Engineer new features
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X[selected_features])
```

### Model Building and Fine-Tuning

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import GridSearchCV
```

```python
# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size=0.2, random_state=42)
```

```python
# Building the regression model
model = RandomForestRegressor()
model.fit(X_train, y_train)
```

```python
# Fine-tuning the model
param_grid = {'n_estimators': [100, 200, 300], 'max_depth': [10, 20, 30]}
```

```python
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='neg_mean_squared_error')
```

```python
grid_search.fit(X_train, y_train)
```

```python
best_model = grid_search.best_estimator_
```

```python
# Evaluate the final model
y_pred = best_model.predict(X_test)
final_mse = mean_squared_error(y_test, y_pred)
```

```python
# Calculating evaluation metrics
mse_test = mean_squared_error(y_test, y_pred)
rmse_test = np.sqrt(mse_test)
r_squared_test = best_model.score(X_test, y_test)

print("Mean Squared Error (MSE) on Test Set:", mse_test)
print("Root Mean Squared Error (RMSE) on Test Set:", rmse_test)
print("R-squared on Test Set:", r_squared_test)
```

#### The model is performing well on the test set, with a low MSE and RMSE, and a high R-squared value. These results indicate that the model is explaining a significant portion of the variance in the target variable (Base MSRP).
